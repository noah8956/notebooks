```matlab
% plot the median sharpness score by location  
% explore_score_spatially.m

clear all;
close all;
clc;

% load FLK_data_gizmo_consolidate_all
load FLK_data_gizmo_exp1_95_scores

X = a_X;
Y = a_Y;
Z = a_Z;
ESD = a_ESD;
max = a_maxscore;

ax = -1250;
ay = -1250;
az = 5000;

bw = 75;
bl = 75;
bh = 750;

minESD = 1;
slices=10000/bh;
boxes=2500/bl;

threshold_bin_size = 1;

median_scores_spatially = zeros(ceil(boxes),ceil(boxes),ceil(slices));
median_scores_spatially_min = zeros(ceil(boxes),ceil(boxes),ceil(slices));

for iz = 1:slices
    for iy = 1:boxes;
        for ix = 1:boxes
            hi = find(X>ax & X<=ax+bw & Y>ay & Y<= ay+bl & Z>az & Z<=az+bh & ESD>minESD);
            median_scores_spatially(ix,iy,iz) = median(a_maxscore(hi));
            median_scores_spatially_min(ix,iy,iz) = median(a_minscore(hi));
            if length(hi) <threshold_bin_size
                median_scores_spatially(ix,iy,iz) = NaN;
                median_scores_spatially_min(ix,iy,iz) = NaN;
            end
            ax = ax+bw;
        end
        ax = -1250;
        ay = ay+bl;
    end
    ay = -1250;
    az = az+bh;
    disp(iz./length(slices))
end

cup = median_scores_spatially;
cup_min = median_scores_spatially_min;

for i=1:9
figure(1)
subplot(3,3,i)
pcolor(cup(:,:,i))
title(['z=',num2str(5000+i*1000), ', n= ',num2str(nanmedian(nanmedian(cup(:,:,i))))])
xlabel('x')
ylabel('y')
colorbar
caxis([0 400])
end

for i=1:9
figure(2)
subplot(3,3,i)
pcolor(cup_min(:,:,i))
title(['z=',num2str(5000+i*1000), ', n= ',num2str(nanmedian(nanmedian(cup_min(:,:,i))))])
xlabel('x')
ylabel('y')
colorbar
caxis([0 4])
end
```
