```matlab
% visualize the nutrient data 
% Redfield ratio CNP -- 106:16:1

clear all; 
close all;

load /Users/noahwalcutt/Documents/MatLab/EN581/consolidate_data_video/EN581_nutrient_data

i=1;

for inisk = 1:length(n_ctd)
    
  hi = find(n_ctd==i);
  plot(n_silicate(hi),-n_z(hi)) 
  xlabel('silicate (uM)')
  ylabel('depth (m)')
  title('EM581 Silicate profiles from Niskins bottles')
  
  hold on
  i=i+1;
  
end

labels = num2str(unique(n_ctd));
legend(labels)
```

```matlab
n_redfield = find(n_ctd ==4);
% Redfield ratio C:Si:N:P = 106:15:16:1

figure(4)
plot(n_silicate(n_redfield)*(16/15),-n_z(n_redfield))
hold on
plot(n_nitrite_nitrate(n_redfield),-n_z(n_redfield))
plot(n_phosphate(n_redfield)*16,-n_z(n_redfield))
% plot(n_ammonium(n_redfield),-n_z(n_redfield))
xlabel('Redfield scaled nutrients')
ylabel('depth (m)')
title('EN581 Redfield scaled nutrients CTD 4')
% title(char(i))

legend('silicate','nitrite nitrate','phosphate')
```
