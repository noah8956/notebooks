```matlab
clear all;
close all;
clc;

load FLK_data_gizmo_exp1

lower = 9000;
step = 200;
upper = lower+step*4;
j=2;
k=1;

for i = 1:4
    
%     hi = find(Z>=10000 & Z<=12000);
    hi = find(Z>=lower & Z<lower+step);
    
    figure(1)
    subplot(4,2,k)
     scatter(Y,Z,'r','filled')
%     axis([-1300 1300 -1300 1300 lower upper])
     axis([-1300 1300 2000 18000])
     title(['Distance = ' num2str(Z(hi(1)))])
     xlabel('DIHM X axis (um)')
     ylabel('Distance from laser (um)')
     hold on
     scatter(Y(hi), Z(hi),'k','.')
     hold off


    subplot(4,2,j)
    histogram2(X(hi),Y(hi),'DisplayStyle','tile')
    axis([-1300 1300 -1300 1300])
    xlabel('DIHM X axis (um)')
    ylabel('DIHM Y axis (um)')
    title(['n = ' num2str(length(hi))])
    
    lower = lower+3000;
    
    j=j+2;
    k=k+2;
end
```
